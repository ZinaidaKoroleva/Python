```python
import math
from sympy import *
from sympy.plotting import plot
x = symbols('x')
y = -12*x**4 - 12*x**3+5*x**2 + 10*x - 30
```

```python
# Определить корни
print(solve(y))
sqrt1,sqrt2 = solve(y)
print(sqrt1,sqrt2)
```

```python
# Найти интервалы, на которых функция возрастает
fx = diff(y)
print(solve(y>0))
```

```python
# Найти интервалы, на которых функция убывает
fx = diff(y)
print(solve(y<0))
```

```python
#Построить график
p = plot(y, (x,-20,20))
```

```python
# Вычислить вершину
print(solve(y))
```

```python
# Определить промежутки, на котором f > 0
print(solve(y>0))
```

```python
# Определить промежутки, на котором f < 0
print(solve(y<0))
```
